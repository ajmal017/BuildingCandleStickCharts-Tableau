```python
import yfinance as yf
```

```python
tesla = yf.download('TSLA', start= '2019-01-01', end ='2020-08-17')
apple = yf.download('AAPL', start= '2019-01-01', end ='2020-08-17')
microsoft = yf.download('MSFT', start= '2019-01-01', end ='2020-08-17')
amazon = yf.download('AMZN', start= '2019-01-01', end ='2020-08-17')
facebook = yf.download('FB', start= '2019-01-01', end ='2020-08-17')
```

```python
[tesla.head(), amazon.head(), facebook.head(),microsoft.head(),apple.head()]
```

```python
amazon.to_csv('Data/amazon.csv', index= False)
tesla.to_csv('Data/tesla.csv', index = False)
apple.to_csv('Data/apple.csv', index = False)
facebook.to_csv('Data/fb.csv', index = False)
microsoft.to_csv('Data/microsoft.csv', index = False)
```

```python
import pandas as pd
```

```python
objects = [amazon, tesla,apple, facebook, microsoft]
for obj in objects:
    obj.reset_index(inplace = True)
```

```python
date = tesla['Date']
```

```python
objects = [amazon, tesla,apple, facebook, microsoft]
all_data = pd.DataFrame(date)
for obj in objects:
    all_data = pd.merge(left = all_data, right = obj, how = 'inner', on = 'Date')
```

```python
cols = all_data.columns
cols
```

```python
new_cols= ['Date','Open_amzn', 'High_amzn', 'Low_amzn', 'Close_amzn', 'Adj_Close_amzn', 'Volume_amzn',
           'Open_tesla', 'High_tesla', 'Low_tesla', 'Close_tesla', 'Adj_Close_tesla', 'Volume_tesla',
           'Open_apple', 'High_apple', 'Low_apple', 'Close_apple', 'Adj_Close_apple', 'Volume_apple',
           'Open_facebook', 'High_facebook', 'Low_facebook', 'Close_facebook', 'Adj_Close_facebook', 'Volume_facebook',
           'Open_microsoft', 'High_microsoft', 'Low_microsoft', 'Close_microsoft', 'Adj_Close_microsoft', 'Volume_microsoft']
```

```python
all_data.columns = new_cols
```
