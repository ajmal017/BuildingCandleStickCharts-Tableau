```python
import yfinance as yf
```

```python
tesla = yf.download('TSLA', start= '2020-01-01', end ='2020-08-17')
apple = yf.download('AAPL', start= '2020-01-01', end ='2020-08-17')
microsoft = yf.download('MSFT', start= '2020-01-01', end ='2020-08-17')
amazon = yf.download('AMZN', start= '2020-01-01', end ='2020-08-17')
facebook = yf.download('FB', start= '2020-01-01', end ='2020-08-17')
```

```python
[tesla.head(), amazon.head(), facebook.head(),microsoft.head(),apple.head()]
```

```python
amazon.to_csv('Data/amazon.csv')
tesla.to_csv('Data/tesla.csv')
apple.to_csv('Data/apple.csv')
facebook.to_csv('Data/fb.csv')
microsoft.to_csv('Data/microsoft.csv')
```

```python
objects = [amazon, tesla,apple, facebook, microsoft]
all_data = pd.DataFrame()
for obj in objects:
    all_data = all_data.merge(left = all_data, right = obj, how = 'inner', on = 'Date')
```

```python
all_data = pd.merge(left = amazon, right= tesla, how = 'inner',on = 'Date')
```
